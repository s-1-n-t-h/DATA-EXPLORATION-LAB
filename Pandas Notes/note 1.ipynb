```python
df = pd.DataFrame({'name':['Alfred','Batman','Catwoman'],
                  'toy':[np.nan,'Batmobile','Bullwhip'],
                  'born':[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
```

```python
df
```

```python
df.dropna() # drops na value containing rows
```

```python
df.dropna(how='all') # drops only if all values are null in 
 # in a row
```

```python
df.dropna(axis=1) # drops all columns having any nulls
```

```python
df.dropna(axis=0) #drops all rows havig any nulls
```

```python
#if you want to drop a row/column that has all values as null
df.dropna(how='all',axis=0)
```

```python
df.dropna(how='all',axis=1)
```

```python
# for dropping duplicates values
df.drop_duplicates(subset=['name'],keep='first')
```

```python
df = pd.DataFrame({'name':['Alfred','Batman','Alfred'],
                  'toy':[np.nan,'Batmobile','Bullwhip'],
                  'born':[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
```

```python
df
```

```python
df.drop_duplicates(subset=['name'],keep='first')
# keeps first copy of duplicate
```

```python
df.drop_duplicates(subset=['name'],keep='last')
# keeps last copy of duplicate
```

```python
df.drop_duplicates(subset=['name'],keep=False)
# keeps no copy of duplicate
```
