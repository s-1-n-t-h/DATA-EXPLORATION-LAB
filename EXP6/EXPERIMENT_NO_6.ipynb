```python
import pandas as pd
```

```python
#preparing the data set
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.drop(['parental level of education','lunch','test preparation course'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#a. Rename method helps to rename column of data frame
df.rename(columns={'gender':'sex'}).head()
```

```python
#b. To rename the column of existing data frame set inplace=True
df.rename(columns={'math score':'ms','reading score':'rs','writing score':'ws'},inplace=True)
```
