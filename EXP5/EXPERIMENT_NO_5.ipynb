```python
#a Creating a Data Frame from scratch

import pandas as pd
```

```python
#using dictionaries
data_set = {
    'product_name':['laptop','printer','tablet','desk','chair'],
    'price':[1200,150,300,450,200],
    'Availability':['Y','N','Y','Y','N']
}
```

```python
data_frame = pd.DataFrame(data_set,index=['p1','p2','p3','p4','p5'])
```

```python
data_frame
```

```python
# using list comprehension

data_set = [
    [1,'Tokyo','Japan',13500],
    [2,'Shangai','China',3920],
    [3,'Mexico City','Mexico',2815],
    [4,'Delhi','India',33578],
    [5,'New York','United States',21483]
]
```

```python
data_frame = pd.DataFrame(data_set,index=['c1','c2','c3','c4','c5'],columns=['Rank','City','Country','Area in km^2'])
```
